<p align="justify">
El origen del Análisis Factorial suele atribuirse a Charles Spearman (1904), en su clásico trabajo sobre inteligencia, donde distingue un factor general (factor G) y cierto número de factores específicos. Hotelling (1933), desarrolló un método de extracción de factores sobre la técnica de "componentes principales".
<br>
<br>
El factorial de un número entero positivo se define como el producto de todos los números naturales anteriores o iguales a él.
Se escribe n!, y se lee "n factorial". (Por definición el factorial de 0 es 1: 0!=1)
<br>
<br>
Por ejemplo, 5! = 5.4.3.2.1 = 120
</p>

# __Modelo de regresión lineal__
****

Se realiza un modelo de regresion lineal con el fin de obtener una prediccion de la valiable de ejecucion de tiempo en Python, esta prediccion nos arroja el tiempo estimado que se demoraria en Cython, teniendo como ejemplo una ejecucion en Python de 1000 segundos, lo que nos arroja un tiempo en Cython de 56.89004, evidenciando el mejoramiento

```python
X = df["Python"]
Y = df["Cython"]
X = X.values.reshape(-1,1)
Y = Y.values.reshape(-1,1)
```

```python
# Se hace la division en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
# Se realiza un modelo de regresion lineal y se entrena
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Se realiza la prediccion de los datos de entrenamiento y los datos de prueba

predictX = modelo.predict(X_train)
predictY = modelo.predict(X_test)
```

```python
# Error cuadratico medio de los datos de entrenamiento
mae1 = mean_squared_error(y_train,predictX)
print("Error de los datos de entrenamiento:" ,  mae1)
```

```python
# Error cuadratico medio de los datos de prueba
mae2 = mean_squared_error(y_test,predictY)
print("Error de los datos de prueba:" ,  mae2)
```

<p align="justify">Como se menciona anteriormente, una vez que ya se cuenta con el modelo de regresión entrenado con el conjunto de datos, se procede a hacer las correspondientes predecciones.
<br>
<br>
Para este caso, el modelo recibirá el tiempo en segundos que le tomó a Python la ejecución del modelo  y este devolverá la predección en segundos del tiempo que le temoria a Cython la resolución del mismo problema.</p>

```python
# Prediccion de un tiempo de 1000 segundos
modelo.predict([[1000]])
```

<p align="justify">De esta manera se obtiene que, lo que le tomaria a Python 1000 segundos a Cython le tomaria casi la mitad con 632.88 segundos. Lo que indica que entre mayor sea la escala del número al que se le quiera obtener el factorial, Cython tendrá un desempeño 100 veces mejor a comparación a Python</p> 

```python
df_predict = pd.DataFrame()
X = []
for i in range(0,100000,2000):  
  X.append(i)
df_predict["Python"] = X
df_predict["Cython"] = modelo.predict(df_predict["Python"].values.reshape(-1,1))
```

```python
# Realizamos una grafica con los resultados obtenidos anteriormente 
# Se evidencia la diferencia de tiempo de los dos experimentos

plt.figure(figsize=(14,8))
plt.scatter(df_predict["Python"],df_predict["Cython"])
plt.title("Tiempo de ejecucion")
plt.xlabel("Tiempo PYTHON")
plt.ylabel("Tiempo CYTHON")
plt.legend()
plt.grid()
```

###__Método de Montecarlo para hacer aproximaciones a PI__
___


<p align="justify">El método de Montecarlo1​ es un método no determinista o estadístico numérico, usado para aproximar expresiones matemáticas complejas y costosas de evaluar con exactitud. El método se llamó así en referencia al Casino de Montecarlo (Mónaco) por ser “la capital del juego de azar”, al ser la ruleta un generador simple de números aleatorios. El nombre y el desarrollo sistemático de los métodos de Montecarlo datan aproximadamente de 1944 y se mejoraron enormemente con el desarrollo de la computadora u ordenador.
<br>
<br>
Dadas estas primeras explicaciones, procederemos a continuación a aplicar el método Montecarlo para estimar el valor de «PI«. La idea aquí consiste en generar de modo aleatorio una serie de puntos (x, y) en un plano 2-D cuadrado de lado 1. Donde a su vez insertaremos un círculo con el mismo diámetro e inscrito en dicho cuadrado. Luego calculamos la proporción de puntos numéricos que se encuentran dentro del círculo y el número total de puntos generados, para, a partir de esos datos, calcular el posible valor de «pi» </p>

<img src = "https://programacionpython80889555.files.wordpress.com/2021/05/estimacion_de_pi_por_montercarlo.gif?w=504"></img>

###__Algoritmo de ordenamiento: Burbuja__
___

<p align="justify">El ordenamiento de burbuja (Bubble Sort en inglés) es un sencillo algoritmo de ordenamiento. Funciona revisando cada elemento de la lista que va a ser ordenada con el siguiente, intercambiándolos de posición si están en el orden equivocado. Es necesario revisar varias veces toda la lista hasta que no se necesiten más intercambios, lo cual significa que la lista está ordenada. Este algoritmo obtiene su nombre de la forma con la que suben por la lista los elementos durante los intercambios, como si fueran pequeñas "burbujas". También es conocido como el método del intercambio directo. Dado que solo usa comparaciones para operar elementos, **se lo considera un algoritmo de comparación, siendo uno de los más sencillos de implementar.**
<br>
<br>
Importante tener en cuenta que de todos los algoritmos de ordenamiento que existen (QuickSort, Shell, Selection) el algoritmo burbuja es el que mas recursos consume, por ende es el mas lento. Es por esto que se escogio, para poner a prueba ambos lenguajes de programación.
</p>

**1.	¿Vale la pena utilizar Cython?**

<p align="justify">
Sí, es importante tener en cuenta que entre mayor sea la adaptación que se haga de código Python a código Cython, mejores serán los resultados obtenidos en cuanto a tiempo de ejecución. Lo anterior se puede observar con los resultados obtenidos en el problema de encontrar un numero factorial y en la implementación de Monte Carlo. Como se observa en las gráficas, ambas curvas siguen estando relativamente cerca una de otra (A pesar de que Cython muestre una mejora en el rendimiento) y esto se debe al cambio realizado de Python a Cython (Ver diferencias .html). De esta manera, se llega a la conclusión de que vale la pena utilizar Cython únicamente si se adapta correctamente el código Python, ya que de no ser así, se estaría desaprovechando los recursos que Cython trae para implementar, de tal manera que solo se estarían ganando unos pocos segundos de ejecución, caso contrario a la implementación desarrollada en el problema de ejecución de matrices.
</p>

<br>


**2.	¿Se pueden obtener mejores resultados?**

<p align="justify">
Con los datos obtenidos se concluye que Cython mejora el rendimiento considerablemente en cuanto a tiempo. Sin embargo, una pregunta que valdría hacerse la pena es ¿Se puede mejorar aún más el rendimiento de nuestras implementaciones y de qué manera?
Durante el desarrollo de este laboratorio, se identificó que va a llegar un punto en donde el algoritmo deja de mejorar, es decir, por más que se mejore el código los tiempos obtenidos van a hacer los mismo y no van a presentar una mejora. Una forma en la que se puede alcanzar un mayor rendimiento de manera relativamente fácil es cambiando la complejidad algorítmica de un problema, por ejemplo, la complejidad algorítmica de los problemas de Monte Carlo y búsqueda con método burbuja es exponencial, es decir O(n) = n^2. De manera que un algoritmo más eficiente seria uno el cual tenga una complejidad algorítmica lineal. Mejorando drásticamente los tiempos de ejecución. 
Otra manera de mejorar el rendimiento de nuestro programa es realizar procesos de paralelización, en el que se haga uso de mas de un nodo computacional. Con el fin de distribuir procesos y aprovechar la memoria distribuida. Una de las herramientas que se pueden utilizar para el proceso de paralelización pueden ser Open MP o MPI.
</p>
<br>
